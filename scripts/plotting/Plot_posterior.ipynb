```python
import pyabc
import parse
import numpy as np

from pyabc.visualization import plot_kde_2d_highlevel
import matplotlib.pyplot as plt
import matplotlib.ticker as mpt

params = {
      #'figure.figsize': (12, 12),
      'legend.fontsize': 20,
      'axes.labelsize': 24,
      'axes.titlesize':24,
      'xtick.labelsize':15,
      'ytick.labelsize':15,
      'text.usetex':True,
     }

plt.rcParams.update(params)

solvers = ['WMM', 'CBM', 'smoldyn']

inference_files = [
    'ss/ss_all_hh/inference_(0.0,0.0,0.0)_All.db',
]


bounds = {
    "chi": (-2, 4),
    "diffusion": (-8, 4),
}



for inf_file in inference_files: 
    db_path = "sqlite:///{}".format(inf_file)
    history = pyabc.History(db_path)
    print(inf_file)
    for model in range(3):
        plot_kde_2d_highlevel(history,
                        y='diffusion', x='chi',
                        m=model,
                        ymin=bounds['diffusion'][0], ymax=bounds['diffusion'][1],
                        xmin=bounds['chi'][0], xmax=bounds['chi'][1],
                        refval={'diffusion': 0, 'chi': 0},
                    )

        
```
